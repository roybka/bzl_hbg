```python
# Importing packages that we'll use to plot

import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
from plotnine import *
from plotnine.data import mpg
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Take a look at a few example rows from our example data-set
mpg.head()
```

```python
# example bar graph


(ggplot(mpg)         # defining what data to use
 + aes(x='class')    # defining what variable to use
 + geom_bar(size=20) # defining the type of plot to use
)
```

```python
# same graph, now save it into file for illustrator

q=(ggplot(mpg)         # defining what data to use
 + aes(x='class')    # defining what variable to use
 + geom_bar(size=20) # defining the type of plot to use
)
q.save('/home/ro/Documents/a.eps')
```

```python
my_x=mpg['class'].value_counts().keys()
my_y=mpg['class'].value_counts().values
```

```python
fig=plt.figure(figsize=(16,9))
plt.bar(my_x,my_y,label='some label')

plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=16)
fig.savefig('/home/ro/Documents/a.eps')

plt.legend()
_=plt.xticks(rotation=45,fontsize=13)
```

```python
fig=plt.figure(figsize=(16,9))
sns.barplot(my_x,my_y,label='some label')

plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=16)
fig.savefig('/home/ro/Documents/a.eps')

plt.legend()
_=plt.xticks(rotation=45,fontsize=13)
```
